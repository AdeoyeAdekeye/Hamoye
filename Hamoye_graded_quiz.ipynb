
Which year had the least correlation with ‘Element Code’?

```python

years = df[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']]
correlations = {}

for column in years.columns:
    correlation = df['Element Code'].corr(df[column])
    correlations[column] = correlation

# Find the year with the least correlation
min_correlation_year = min(correlations, key=correlations.get)
min_correlation = correlations[min_correlation_year]

print(f"The year with the least correlation is {min_correlation_year} with a correlation of {min_correlation}")
```

What is the total sum of Wine produced in 2015 and 2018 respectively?

```python



# Calculate the total sum of Wine produced in 2015 and 2018
wine_production_2015 = df['Y2015'][df['Item'] == 'Wine'].sum()
wine_production_2018 = df['Y2018'][df['Item'] == 'Wine'].sum()

# Print the total sum of Wine produced in 2015 and 2018
print(f"Total Wine production in 2015: {wine_production_2015}")
print(f"Total Wine production in 2018: {wine_production_2018}")








```

What is the mean and standard deviation across the whole dataset for the year 2017 to 2 decimal places?

```python
# Extract the column for the year 2017
year_2017 = df['Y2017']

# Calculate the mean and standard deviation
mean_2017 = round(year_2017.mean(), 2)  # Round to two decimal places
std_2017 = round(year_2017.std(), 2)    # Round to two decimal places

# Print the results
print(f"Mean for the year 2017: {mean_2017}")
print(f"Standard Deviation for the year 2017: {std_2017}")
```

Perform a groupby operation on ‘Element’.  What year has the highest sum of Stock Variation?



```python

# Filter the DataFrame to only include rows where 'Element' is 'Stock Variation'
filtered_df = df[df['Element'] == 'Stock Variation']

#calculate the sum of 'Stock Variation' for each year
grouped = filtered_df.sum()


# Print the result
print(f"The year with the highest sum of Stock Variation is {grouped}")
```

What is the total number and percentage of missing data in 2014 to 3 decimal places?

```python

filtered_df = df['Y2014']

# Calculate the total number of missing values in the filtered DataFrame
total_missing = filtered_df.isnull().sum().sum()

# Calculate the total number of values in the filtered DataFrame
total_values = filtered_df.size

# Calculate the percentage of missing data to three decimal places
percentage_missing = (total_missing / total_values) * 100

# Print the results
print(f"Total Missing Data in 2014: {total_missing}")
print(f"Percentage of Missing Data in 2014: {percentage_missing:.3f}%")
```

Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the 7th lowest sum in 2017?

```python


# Select the columns 'Y2017' and 'Area'
selected_columns = df[['Y2017', 'Area']]

# Group the DataFrame by 'Area' and calculate the sum of 'Y2017' for each Area
grouped = selected_columns.groupby('Area')['Y2017'].sum().reset_index()

# Sort the DataFrame by the sum of 'Y2017' in ascending order
sorted_df = grouped.sort_values(by='Y2017')

# Find the 7th lowest sum in 'Y2017'
seventh_lowest_sum = sorted_df['Y2017'].iloc[6]

# Find the Area(s) with the 7th lowest sum in 'Y2017'
areas_with_seventh_lowest_sum = sorted_df[sorted_df['Y2017'] == seventh_lowest_sum]

# Print the result
print("Areas with the 7th Lowest Sum in Y2017:")
print(areas_with_seventh_lowest_sum)
```

Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the highest sum in 2017?

```python


# Select the columns 'Y2017' and 'Area'
selected_columns = df[['Y2017', 'Area']]

# Group the DataFrame by 'Area' and calculate the sum of 'Y2017' for each Area
grouped = selected_columns.groupby('Area')['Y2017'].sum().reset_index()

# Find the Area with the highest sum in 'Y2017'
area_with_highest_sum = grouped.loc[grouped['Y2017'].idxmax()]

# Print the result
print("Area with the Highest Sum in Y2017:")
print(area_with_highest_sum)
```
